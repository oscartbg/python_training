## Area Pots (Series/Dataframe) <a id="12"></a>

**Questions:** Bagaimana trend jumlah kriminal dari 3 borough di tahun 2008-2016?

```python
# data kriminal di 3 borough

Westminster=criminal[(criminal['borough'] == 'Westminster')].groupby(["year"])["value"].sum().reset_index(name ='Westminster').set_index('year')
Lambeth=criminal[(criminal['borough'] == 'Lambeth')].groupby(["year"])["value"].sum().reset_index(name ='Lambeth').set_index('year')
Southwark=criminal[(criminal['borough'] == 'Southwark')].groupby(["year"])["value"].sum().reset_index(name ='Southwark').set_index('year')

# gabung jadi satu
result = pd.concat([Westminster, Lambeth,Southwark], axis=1)
result
```

```python
# plot area 
result.plot(kind='area', 
            stacked=False,
            figsize=(12,6),
            alpha=0.3)

plt.title('Perbandingan kriminal di 3 borough') # add a title to the area plot
plt.ylabel('Jumlah Kriminal') # add y-label
plt.xlabel('Tahun') # add x-label

plt.show()
```

**Insight:** Pada borough Lambeth,Southwark trend kriminalnya cukup landai tidak ada perubahan yang cukup signifikan di tiap tahunnya, sedangkan di Westminster mengalami peningkatan jumlah kriminalitas yang signifikan pada tahun 2012

## Pie Charts <a id="6"></a>

**Question:** Jenis kriminal apa yang paling banyak terjadi di borough Westminster di tahun 2012

**Insight:** Theft and Handling mendominasi kriminal di Westminster tahun 2012

## Word Clouds <a id="8"></a>


```python

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
text =""
max_words=30000
data_borough =criminal.groupby(["borough"])["value"].sum().reset_index(name ='total_kriminal').set_index('borough')

total_kriminal = data_borough['total_kriminal'].sum()
```

```python
# hitung berapa kali borough akan muncul

for borough in data_borough.index:
    borough_name = borough
    for s in " ,()-'":
        borough_name = "".join(borough_name.split(s))
    word_count= int(data_borough.loc[borough,"total_kriminal"] / float(total_kriminal) * max_words)
    text = text + ((borough_name + " ") * word_count)

print("panjang text: ", len(text))
```

```python
# instantiate a word cloud object
your_wordcloud = WordCloud(
    background_color='white',
    max_words=max_words,
    collocations= False,
    stopwords=stopwords
)

# generate the word cloud
your_wordcloud.generate(text)
```

```python
fig = plt.figure(figsize=(15,6))
plt.imshow(your_wordcloud, interpolation='bilinear')

plt.axis('off')
plt.show()
```

## Folium

```python
import folium
from folium import plugins
```

```python
# koordinat London yang menjadi pusat map
london_coordinate = [51.509865,-0.118092]

# data jumlah kriminal di setiap borough
df2=criminal.groupby(["borough"])["value"].sum()
```

```python
#gabung dengan peta pembagian london borough supaya bisa mendapatkan heatmap kriminalitas di tiap borough
london_geo = r'https://raw.githubusercontent.com/radoi90/housequest-data/master/london_boroughs.geojson' #peta london borough
london_map=folium.Map(location=london_coordinate,zoom_start=10)

london_map.choropleth(
    geo_data=london_geo,
    name='choropleth',
    data=df2,
    columns=['borough', 'value'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime rate in boroughs of london'
    
)

# display london map
display(london_map)
```
